```python
# 6-1 구구단 2단 출력하기
def gugu_2(num):
  result=[]
  i=1
  while i < 10:
    result.append(i*num)
    i+=1
  return result
print(gugu_2(2))
```

```python
# 6-2 1000사이의 숫자 중에 3과 5의 배수 구하기
result = 0
for i in range(1,1000):
  if i % 3 == 0 or i % 5 == 0:
    result+=i
print(result)
```

```python
# 6-3 게시판 페이징하기
def getTotalPage(m,n):
  if m % n == 0:
    return m // n
  else:
    return (m // n) + 1

print(getTotalPage(5,10))
print(getTotalPage(15,10))
print(getTotalPage(25,10))
print(getTotalPage(30,10))
```

```python
# 6-4 메모장 기능 만들기(cmd이용)
# pythjon memo.py - a "Life is too short"
# type memo.txt
# python memo.py -v
import sys

option = sys.argv[1]

if option =='-a':
    memo=sys.argv[2]
    f=open('memo.txt','a')
    f.write(memo)
    f.write('\n')
    f.close()
elif option =='-v':
    f=open('memo.txt')
    memo=f.read()
    f.close()
    print(memo)
```

```python
# 6-5 탭을 4개의 공백으로 바꾸기
# python tabto4.py a.txt b.txt 탭을 4개의 공백으로 바꾸어 b.txt에 저장
import sys

src = sys.argv[1]
dst = sys.argv[2]

# a.txt열고 읽기
f=open(src)
tab_content=f.read()
f.close()

# 탭을 공백 4개로 바꾸기
space_content=tab_content.replace('\t'," "*4)

# b.txt에 저장
f=open(dst,'w')
f.write(space_content)
f.close()
```

```python
# 6-6 하위 디렉터리 검색하기
import os

def search(dirname):
    try:
        filenames = os.listdir(dirname)
        for filename in filenames:
            full_filename = os.path.join(dirname,filename)
            if os.path.isdir(full_filename):
                search(full_filename)
        else:
            ext = os.path.splitext(full_filename)[-1]
            if ext =='.py':
                print(full_filename)
    except PermissionError:
        pass

search("C:/")
```
