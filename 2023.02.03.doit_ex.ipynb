```python
# 5★
def fibo(n):
  if n ==0:
    return 0
  elif n ==1:
    return 1
  else:
    return fibo(n-2)+fibo(n-1)

for i in range(10):
  print(fibo(i),end=' ')
```

```python
# 8★
f = open('abc.txt','r')
lines = f.readlines() #모든 라인을 읽음
f.close()

lines.reverse() # 역순 정렬

f = open('abc.txt','w')
for line in lines:
    line = line.strip() #줄 바꿈 문자 제거
    f.write(line)
    f.write('\n') # 줄 바꿈 문자 삽입
f.close()
```

```python
# 9★
f = open("sample.txt")
lines = f.readlines()
f.close()

total = 0
for line in lines:
  score = int(line)
  total+=score

average = total/len(lines)

f = open("result.txt","w")
f.write(str(average))
f.close()
```

```python
# 10
class Calculator:
    def __init__(self,numberlist):
      self.numberlist = numberlist

    def sum(self):
      total=0
      for i in self.numberlist:
        total+=i
      return total

    def avg(self):
        total = self.sum()
        return total / len(self.numberlist)
        
cal1 = Calculator([1,2,3,4,5])
cal1.sum()
cal1.avg()

cal2 = Calculator([6,7,8,9,10])
cal2.sum()
cal2.avg()
```
