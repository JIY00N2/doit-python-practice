```python
# 3-1
def is_odd(number):
  if number % 2 == 1:
      return True
  else:
      return False

print(is_odd(3))
print(is_odd(4))
```

```python
# 3-2
def avg_numbers(*args):
  result=0
  for i in args:
    result+=i
  return result/len(args)

print(avg_numbers(1,2))
print(avg_numbers(1,2,3,4,5))
```
